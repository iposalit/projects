```python
import Bio
from Bio import SeqIO, SearchIO, Entrez
from Bio.Seq import Seq
from Bio.SeqUtils import GC
from Bio.Blast import NCBIWWW
from Bio.Data import CodonTable
import pandas as pd
import numpy as np
```

```python
Entrez.email = "cschulte@hsph.harvard.edu"
```

```python
genes = pd.read_csv('dataset_filtered.csv')
info = pd.read_csv('datainfo_filtered.csv')
```

```python
info
```

```python
accList = []
for accession in info["genomeAcc"]:
  try:
    accession = accession.split(", ")
  except:
    continue

  for seq in accession:
    if seq in accList:
      continue
    else:
      accList.append(seq)
      fetch_handle = Entrez.efetch("nuccore", id=seq, rettype="gbwithparts", retmode="text")

      with open(seq + ".gbk", "w") as output_handle:
        output_handle.write(fetch_handle.read())
    
      print("Saved %s.gbk" % seq)
```

```python
def get_cds_feature_with_qualifier_value(seq_record, name, value):
    """Function to look for CDS feature by annotation value in sequence record.
    
    e.g. You can use this for finding features by locus tag, gene ID, or protein ID.
    """
    for feature in genome_record.features:
        if feature.type == "CDS" and value in feature.qualifiers.get(name, []):
            return feature
            
    return "None"
```

```python
cds_dict ={}

for acc in accList:
  print(acc)
  cds_count = 0
  genome_record = SeqIO.read(acc + ".gbk", 'genbank')
  for feature in genome_record.features:
    if feature.type == "CDS":
      cds_count += 1
  cds_dict[acc] = cds_count
```

```python
genes["gene_sequence"] = 0
genes["protein_sequence"] = 0
genes["strand"] = 0
taxID_tmp = 0

for i in range(len(genes)):
  if i%1000 == 0:
    print(str(round(i/len(genes)*100,2)) + "% done")
  cds_feature = "None"
  taxID = genes.at[i, "taxaID"]
  if taxID != taxID_tmp:
    print(taxID)
    taxID_tmp = taxID
    try:
      seqID = list(info[info['taxID'] == taxID]["genomeAcc"].str.split(r", "))[0][0]
    except:
      continue
    file = seqID + '.gbk'
    print(file)
    genome_record = SeqIO.read(file, 'genbank')
  
  gene = genes.at[i, "gene"]
  locus = genes.at[i, "locus"]

  cds_feature = get_cds_feature_with_qualifier_value(genome_record, "locus_tag", gene)
  if cds_feature == "None":
    cds_feature = get_cds_feature_with_qualifier_value(genome_record, "old_locus_tag", gene)
    if cds_feature == "None":
      cds_feature = get_cds_feature_with_qualifier_value(genome_record, "locus_tag", locus)
      if cds_feature == "None":
        cds_feature = get_cds_feature_with_qualifier_value(genome_record, "old_locus_tag", locus)
  if cds_feature != "None":
    geneSeq = cds_feature.extract(genome_record.seq)
    try:
      proteinSeq = geneSeq.translate(table=11, cds=True)
    except:
      proteinSeq = 0
    genes.loc[i, "gene_sequence"] = str(geneSeq)
    genes.loc[i, "protein_sequence"] = str(proteinSeq)
    genes.loc[i, "strand"] = cds_feature.strand
    if i%1000 == 0:
      print(str(round(np.sum(genes["gene_sequence"] != 0)/len(genes)*100,2)) + "% succesfully matched")
```
